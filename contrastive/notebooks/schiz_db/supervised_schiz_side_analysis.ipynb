```python
# load subsets
train = pd.read_csv('/neurospin/dico/data/deep_folding/current/datasets/schiz/train_subjects.csv', names=['Subject'])
val = pd.read_csv('/neurospin/dico/data/deep_folding/current/datasets/schiz/val_subjects.csv', names=['Subject'])
test_intra = pd.read_csv('/neurospin/dico/data/deep_folding/current/datasets/schiz/test_intra_subjects.csv', names=['Subject'])
test = pd.read_csv('/neurospin/dico/data/deep_folding/current/datasets/schiz/test_subjects.csv', names=['Subject'])

preds['subset'] = ['' for i in range(preds.shape[0])]
preds.loc[preds.Subject.isin(train.Subject), 'subset'] = 'train'
preds.loc[preds.Subject.isin(val.Subject), 'subset'] = 'val'
preds.loc[preds.Subject.isin(test_intra.Subject), 'subset'] = 'test_intra'
preds.loc[preds.Subject.isin(test.Subject), 'subset'] = 'test'
preds.subset.unique()

# test intra pas pris en compte par embeddings pipeline en fait
```
